```python
%matplotlib inline

import os
import dicom
import math
import matplotlib.pyplot as plt

root_dir = "/Users/Breakend/Documents/datasets/sciencebowl2015"

def get_frames(root_path):
    """Get path to all the frame in view SAX and contain complete frames"""
    ret = {}
    for root, _, files in os.walk(root_path):
        if len(files) == 0 or not files[0].endswith(".dcm") or root.find("sax") == -1:
           continue
        prefix = files[0].rsplit('-', 1)[0]
        data_index = int(root.rsplit('/', 3)[1])
        fileset = set(files)
        expected = ["%s-%04d.dcm" % (prefix, i + 1) for i in range(30)]
        if all(x in fileset for x in expected):
            if data_index in ret:
               ret[data_index].append([root + "/" + x for x in expected])
            else:
               ret[data_index] = [[root + "/" + x for x in expected]]
        else:
            backup_expected = ["%s-%04d-0002.dcm" % (files[0].rsplit('-', 2)[0], i + 1) for i in range(30)]
            if all(x in fileset for x in backup_expected):
                if data_index in ret:
                   ret[data_index].append([root + "/" + x for x in backup_expected])
                else:
                   ret[data_index] = [[root + "/" + x for x in backup_expected]]
    return ret
print("starting")
slice_distances_from_middle = {}
for patient in get_frames(root_dir).values():
    slice_locations = []
    for sax_slice in patient:
        loc = dicom.read_file(sax_slice[0]).SliceLocation
        slice_locations.append(loc)
    slice_locations.sort()
    middle = int(slice_locations[len(slice_locations)/2])
    distance_from_middle = [math.sqrt((middle - int(x))**2) for x in slice_locations]
    for distance in distance_from_middle:
        if distance in slice_distances_from_middle:
            slice_distances_from_middle[distance] += 1
        else:
            slice_distances_from_middle[distance] = 1
print(slice_distances_from_middle)
```

```python
plt.bar(range(len(slice_distances_from_middle)), slice_distances_from_middle.values(), align='center')
plt.xticks(range(len(slice_distances_from_middle)), slice_distances_from_middle.keys())
plt.show()
```
